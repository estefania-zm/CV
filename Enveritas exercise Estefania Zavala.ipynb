## Task 1

```python
task_1
```

```python
task_1.groupby('Region')['Yield (kg green per hectare)'].mean()
```

```python
# Average yield per region
avg_yield_per_region = task_1.groupby('Region')['Yield (kg green per hectare)'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Yield (kg green per hectare)', data=avg_yield_per_region, palette='viridis')
plt.title('Average Yield per Region')
plt.xlabel('Region')
plt.ylabel('Average Yield (kg green per hectare)')
plt.xticks(rotation=45)  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Coffee form sold', y='Yield (kg green per hectare)', data=task_1, palette='Set2')
plt.title('Yield vs Coffee Form Sold')
plt.xlabel('Coffee Form Sold')
plt.ylabel('Yield (kg green per hectare)')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
# Applying log transformation to make it easier to read
task_1['Log Yield'] = np.log1p(task_1['Yield (kg green per hectare)'])

# Plot with log-transformed data
plt.figure(figsize=(10, 6))
sns.boxplot(x='Coffee form sold', y='Log Yield', data=task_1, palette='Set2')
plt.title('Log-Transformed Yield vs Coffee Form Sold')
plt.xlabel('Coffee Form Sold')
plt.ylabel('Log Yield (kg green per hectare)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Intercrop food tree with coffee', hue='Coffee Region', data=task_3, palette='viridis')
plt.title('Intercrop food tree with coffee')
plt.xlabel('Intercrop food tree with coffee')
plt.ylabel('Count')
plt.legend(title='Region')
plt.show()
```

```python
# Rows where pruning techniques are not empty
pruning_data = task_3[task_3['Pruning techniques used'].notnull()]

# Counting the occurrences of each pruning technique
pruning_techniques = pruning_data['Pruning techniques used'].str.split(' ', expand=True).stack().value_counts()

# Plot
plt.figure(figsize=(10, 6))
pruning_techniques.plot(kind='bar', color='skyblue')
plt.title('Most Common Pruning Techniques Used')
plt.ylabel('Frequency')
plt.xlabel('Pruning Technique')
plt.show()
```

```python
plt.figure(figsize=(16, 10)) 
sns.boxplot(x='Pruning techniques used', y='Yield (kg/green) per hectare', data=task_3)
plt.title('Yield vs Pruning Techniques')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.countplot(x='Coffee Region', hue='Pruning techniques used', data=task_3)
plt.title('Pruning Techniques by Region')
plt.xticks(rotation=45)
plt.show()

```
