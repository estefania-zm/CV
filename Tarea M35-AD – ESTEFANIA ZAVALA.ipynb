Importar la librería NumPy.  Generar estadística descriptiva básica en las columnas unit_price y quantity: Cálculo de la media, mediana, moda

```python
import numpy as np
```

```python
#Validación de columnas numéricas
df.dtypes
```

```python
#Transformar columnas a Array Numpy
unit_price = df[['Unit price']].to_numpy()
quantity = df[['Quantity']].to_numpy()
```

```python
#Cálculos unit_price
print(f'unit_price mean: {np.mean(unit_price)}')
print(f'unit_price median: {np.median(unit_price)}')
up_val, up_count = np.unique(unit_price, return_counts=True)
up_index = np.argmax(up_count)
print(f'quantity mode: {up_val[up_index]}')

#Cálculos quantity
print(f'quantity mean: {np.mean(quantity)}')
print(f'quantity median: {np.median(quantity)}')
qt_val, qt_count = np.unique(quantity, return_counts=True)
qt_index = np.argmax(qt_count)
print(f'quantity mode: {qt_val[qt_index]}')
```

Obtener el ticket promedio:
1) por ciudad,

```python
#Cantidad de ciudades en el Df
df['City'].unique()
```

```python
df.groupby('City')['Total'].mean().rename('Ticket promedio').reset_index()
```

2) por product line y ciudad,

```python
#Cantidad de lineas de productos en el Df
df['Product line'].unique()
```

```python
df.groupby(['City','Product line'])['Total'].mean().rename('Ticket promedio').reset_index()
```

3) por género,

```python
df.groupby('Gender')['Total'].mean().rename('Ticket promedio').reset_index()
```

4) por género y product line

```python
df.groupby(['Gender','Product line'])['Total'].mean().rename('Ticket promedio').reset_index()
```

Usar el método rank para generar top 5 de:
1) ventas por ciudad,

```python
df['top_ventas'] = df.groupby('City')['Total'].rank(ascending=False)
df[['City','Total','top_ventas']].loc[(df['top_ventas'] < 6)].sort_values(['City','top_ventas'])
```

2) ventas por member,

```python
members = df.loc[(df['Customer type'] == 'Member')]
members['top_member'] = members['Total'].rank(ascending=False)
members[['Customer type','Total','top_member']].loc[(members['top_member'] < 6)].sort_values('top_member')
```

3) ventas por payment.  Obtener además el % de aporte de cada categoría. 

```python
df['top_payment'] = df.groupby('Payment')['Total'].rank(ascending=False)
df[['Payment','Total','top_payment']].loc[(df['top_payment'] < 6)].sort_values(['Payment','top_payment'])
```

P.ej. La ciudad X aporta con el 25% del total de ventas, etc.

```python
ventas_totales = df.groupby('City')['Total'].sum()
porcentaje_ventas = (ventas_totales / ventas_totales.sum()) * 100
result = pd.DataFrame({'Ventas Totales': ventas_totales, 'Porcentaje': porcentaje_ventas})
print(result)
```

```python
total_payment = df.groupby('Payment')['Total'].sum()
porcentaje_payment = (total_payment / total_payment.sum()) * 100
perc_pay = pd.DataFrame({'Ventas por método de pago': total_payment, 'Porcentaje': porcentaje_payment})
print(perc_pay)
```
