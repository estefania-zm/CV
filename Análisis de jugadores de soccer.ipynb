- Leer archivo fifa_eda.csv  

```python
import pandas as pd
import numpy as np
import plotly.express as px
import os
```

```python
os.chdir('C:/Users/estef/Desktop/data')
df = pd.read_csv('fifa_eda.csv')
df.sample(5)
```

- Elemento 1: Obtener matriz de correlación y dejarla como heat map 

```python
def ColNum(x):
    numeric_columns = x.select_dtypes(include=['number'])
    return numeric_columns

df_num = ColNum(df)
```

```python
df_num.corr()
```

```python
px.imshow(df_num.corr(),text_auto=True)
```

- Elemento 2: Gráfico de Correlación que responda – “Cuál es la relación entre la edad y el overall?” – explicar la salida 

```python
overall_age = df[['Overall','Age']]
overall_age.corr()
```

```python
px.imshow(overall_age.corr(),text_auto=True)
```

```python
px.histogram(overall_age, x='Age', y = 'Overall')
```

```python
px.strip(overall_age, x='Age', y = 'Overall')
```

Con los datos obtenidos se puede apreciar que existe una correlación entre la edad de los jugadores y su ranking en la lista de jugadores, presentandose un aumento en la cantidad de jugadores con un ranking alto entre los 21 y los 26 años.
Aunado a lo mencionado antes, se observa que esta correlación no es la regla ya que hay tambien jugadores por debajo y por encima de ese rango de edad que mantienen un puesto alto en el ranking.

- Elemento 3: Generar un gráfico de barras por club que indique el número de jugadores 

```python
#df[['Overall','Age']]
#pd.DataFrame(df.groupby('Club').size())
club_players = df.groupby('Club').size().rename('Players').reset_index().sort_values('Players')
club_players
```

```python
px.bar(club_players,x='Club',y='Players')
```

- Elemento 4: Hacer un gráfico multipanel que indique la relación altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho 

```python
px.strip(df,x = 'Skill Moves',y = 'Height',facet_col='Preferred Foot')
```

- Elemento 5: Un gráfico que crea importante para mostrar a los potenciales cracks

```python
df2 = df[['Name','Club','Joined','Nationality','Age','Weight','Height','Skill Moves','Overall']].sort_values('Overall', ascending=False)
df2
```

```python
def ColNum(x):
    numeric_columns = x.select_dtypes(include=['number'])
    return numeric_columns

df2_num = ColNum(df2)
px.imshow(df2_num.corr(),text_auto=True)
```

```python
px.scatter(df2, x='Nationality', y = 'Overall')
```

```python
px.scatter(df2, x='Club', y = 'Overall')
```

```python
nuevos = df2[df2['Age'] < 21 ]
nuevos.sort_values('Skill Moves', ascending = False)
```

```python
nuevos_num = ColNum(nuevos)
px.imshow(nuevos_num.corr(),text_auto=True)
```

```python
px.scatter(nuevos, x='Joined', y = 'Overall',color='Age')
```

Con este gráfico se pueden apreciar los jugadores que a pesar de su corta edad y del poco tiempo que llevan en sus equipos, han logrado destacar en el ranking.
Tambien se podría tomar en cuenta el potencial de los jugadores que entraron hace ya varios años, que aún son jovenes y estan a punto de entrar a la edad en la que llegan muchos de los jugadores mas destacados a tener su mejor rendimiento.

```python
px.bar(nuevos.groupby('Club').size().rename('Players').reset_index().sort_values('Players'),x='Club',y='Players')
```
